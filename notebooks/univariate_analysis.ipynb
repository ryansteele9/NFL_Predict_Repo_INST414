```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/ryansteele/my_repo2/data/processed/matchups/matchups_all_seasons.csv")

df.head()
```

```python
important_features = [
    'point_diff', 'target_next_point_diff', 'rolling_points_for_3',
    'rolling_point_diff_3', 'rolling_turnover_diff_3',
    'rolling_win_rate_5', 'offensiveyards', 'turnover_diff'
]

df[important_features].describe().T
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['target_next_point_diff'], kde=True, bins=30)
plt.title("Distribution of Next Game Point Differential")
plt.xlabel("Next Game Point Diff")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rolling_points_for_3'], kde=True, bins=30)
plt.title("Rolling 3-Game Average Points Scored")
plt.xlabel("Rolling Points (3 games)")
plt.ylabel("Frequency")
plt.show()
```
